```python

!pip install pandas transformers keybert plotly sentence-transformers umap-learn -q


import pandas as pd
import re
import numpy as np
import torch
import plotly.express as px
from transformers import AutoModelForSeq2SeqLM, NllbTokenizer
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
import umap


df = pd.read_csv("telegram_messages.csv")

def clean_hebrew_text(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\u0590-\u05FF\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['clean_message'] = df['message'].apply(clean_hebrew_text)
texts = df['clean_message'].dropna()
texts = texts[texts.str.strip() != ""]

sampled_texts = texts.sample(n=50000, random_state=42).tolist()

model_name = "facebook/nllb-200-distilled-600M"
tokenizer = NllbTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

src_lang = "heb_Latn"
tgt_lang = "eng_Latn"

def translate_with_nllb_batch(texts, batch_size=128, max_length=256):
    translations = []
    tokenizer.src_lang = src_lang
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        encoded = tokenizer(batch_texts, return_tensors="pt", padding=True, truncation=True, max_length=max_length).to(device)
        generated_tokens = model.generate(
            **encoded,
            forced_bos_token_id=tokenizer.convert_tokens_to_ids(tgt_lang),
            max_length=max_length
        )
        decoded = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
        translations.extend(decoded)

    return translations

translated_texts = translate_with_nllb_batch(sampled_texts)

sbert_model = SentenceTransformer("all-MiniLM-L6-v2")
kw_model = KeyBERT(sbert_model)

topics = kw_model.extract_keywords(
    translated_texts,
    keyphrase_ngram_range=(1, 3),
    stop_words='english',
    top_n=1
)
```

```python
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer
import umap
import pandas as pd
import plotly.express as px


all_keywords = [kw for doc in topics if doc for kw in doc]
keyword_texts = [kw[0] for kw in all_keywords]

keyword_counts = pd.Series(keyword_texts).value_counts().reset_index()
keyword_counts.columns = ["Topic", "Count"]


embeddings = sbert_model.encode(keyword_counts["Topic"].tolist())


n_clusters = 30
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(embeddings)
keyword_counts["Cluster"] = clusters

reducer = umap.UMAP(n_neighbors=10, min_dist=0.3, metric='cosine', random_state=42)
embedding_2d = reducer.fit_transform(embeddings)
keyword_counts["x"] = embedding_2d[:, 0]
keyword_counts["y"] = embedding_2d[:, 1]

representatives = keyword_counts.groupby("Cluster").apply(lambda g: g.nlargest(1, "Count")).reset_index(drop=True)

fig = px.scatter(
    representatives,
    x="x",
    y="y",
    size="Count",
    text="Topic",
    color="Cluster",
    title="Broader Semantic Topics (Clustered & Aggregated)",
    size_max=60
)
fig.update_traces(textposition='top center')
fig.show()
```

```python

!pip install transformers sentencepiece tqdm -q

import pandas as pd
import re
from tqdm.notebook import tqdm
import torch
from transformers import MarianMTModel, MarianTokenizer

model_name = 'Helsinki-NLP/opus-mt-ar-en'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)


def clean_arabic_text(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\u0600-\u06FF\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

israel_keywords = [
    "إسرائيل", "الاحتلال", "الصهاينة", "القدس", "تل أبيب", "فلسطين المحتلة",
    "عكا", "يافا", "حيفا", "بئر السبع", "نتانيا", "إيلات", "اشدود", "صاروخ", "صفارات"
]

def is_about_israel(text):
    return any(keyword in text for keyword in israel_keywords)


def translate_batch(texts, batch_size=16):
    results = []
    for i in tqdm(range(0, len(texts), batch_size), desc="Translating"):
        batch_texts = texts[i:i+batch_size]
        tokens = tokenizer(batch_texts, return_tensors="pt", padding=True, truncation=True)
        tokens = {k: v.to(device) for k, v in tokens.items()}
        with torch.no_grad():
            generated = model.generate(**tokens, max_length=512)
        decoded = [tokenizer.decode(g, skip_special_tokens=True) for g in generated]
        results.extend(decoded)
    return results

df = pd.read_csv("arabic_messages.csv")

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df = df[df['date'].dt.year >= 2019].dropna(subset=['message'])

df = df.sample(n=min(10000, len(df)), random_state=42).reset_index(drop=True)
print(f"🗓️ Using {len(df)} messages from 2019 and later")

df['clean_message'] = df['message'].apply(clean_arabic_text)
df['is_about_israel'] = df['clean_message'].apply(is_about_israel)

df_israel = df[df['is_about_israel']].copy()
print(f"🇮🇱 Found {len(df_israel)} Israel-related messages")

unique_texts = df_israel['clean_message'].dropna().unique().tolist()
translated_texts = translate_batch(unique_texts)

translation_map = dict(zip(unique_texts, translated_texts))
df_israel['translated_message'] = df_israel['clean_message'].map(translation_map)

df_israel.to_csv("telegram_messages_israel_2019plus_translated.csv", index=False)
print("✅ Saved to 'telegram_messages_israel_2019plus_translated.csv'")
```

```python
# !pip install pandas folium geopy transformers sentencepiece plotly -q

import pandas as pd
import re
from collections import Counter, defaultdict
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import plotly.graph_objects as go

df = pd.read_csv("telegram_messages.csv")

def clean_hebrew_text(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\u0590-\u05FF\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['clean_message'] = df['message'].apply(clean_hebrew_text)

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

def translate_words(words):
    inputs = tokenizer(
        [f"__heb_Hebr__ {w}" for w in words],
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=32
    )
    forced_bos_token_id = tokenizer.convert_tokens_to_ids("__eng_Latn__")
    translated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=forced_bos_token_id
    )
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)

hebrew_country_map = {
    "ישראל", "רוסיה", "אוקראינה", "איראן", "עזה", "מצרים", "לבנון", "סוריה", "ירדן", "תימן", "סעודיה", "קטאר", "עיראק",
    "טורקיה", "צרפת", "בריטניה", "גרמניה", "ארצות הברית", "סין", "הודו", "אתיופיה", "קנדה", "אוסטרליה", "פולין",
    "צ'כיה", "הולנד", "שוודיה", "שוויץ", "בלגיה", "ספרד", "פורטוגל", "יפן", "פנמה", "צ'ילה", "ארגנטינה", "ברזיל", "מרוקו"
}

def find_hebrew_countries(text):
    return list({country for country in hebrew_country_map if country in text})

df['hebrew_countries'] = df['clean_message'].apply(find_hebrew_countries)
df['hebrew_countries'] = df['hebrew_countries'].apply(lambda x: list(set(x)))

df_pairs = df[df['hebrew_countries'].apply(lambda x: isinstance(x, list) and len(x) == 2)].copy()

unique_hebrew_countries = sorted({c for lst in df_pairs['hebrew_countries'] for c in lst})
translated_countries = translate_words(unique_hebrew_countries)
hebrew_to_english = dict(zip(unique_hebrew_countries, translated_countries))

df_pairs['english_countries'] = df_pairs['hebrew_countries'].apply(
    lambda lst: [hebrew_to_english.get(c, c).strip() for c in lst]
)

sentiment_pipeline = pipeline("sentiment-analysis", model="avichr/heBERT_sentiment_analysis")

def is_valid_country(name):
    return isinstance(name, str) and name.strip() and name.replace(" ", "").isalpha()

pair_counts = Counter()
pair_sentiments = defaultdict(list)

for _, row in df_pairs.iterrows():
    pair = row['english_countries']
    if isinstance(pair, list) and len(pair) == 2:
        c1, c2 = sorted([p.strip() for p in pair])
        if is_valid_country(c1) and is_valid_country(c2):
            text = row['clean_message']
            tokens = sentiment_pipeline.tokenizer.tokenize(text)
            if len(tokens) > 510:
                tokens = tokens[:510]
                text = sentiment_pipeline.tokenizer.convert_tokens_to_string(tokens)

            sentiment = sentiment_pipeline(text)[0]
            label = sentiment['label']
            pair_counts[(c1, c2)] += 1
            pair_sentiments[(c1, c2)].append(label)

from collections import Counter as LabelCounter

def majority_label(label_list):
    return LabelCounter(label_list).most_common(1)[0][0]

cleaned_pairs = [
    (pair[0], pair[1], pair_counts[pair], majority_label(pair_sentiments[pair]))
    for pair in pair_counts
    if len(pair) == 2
]

pair_df = pd.DataFrame(
    cleaned_pairs,
    columns=["First Country", "Second Country", "Message Count", "Sentiment Label"]
).sort_values(by="Message Count", ascending=False).reset_index(drop=True)

geolocator = Nominatim(user_agent="geoapi")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

all_countries = {c for pair in pair_counts for c in pair}
country_coords = {}
for country in all_countries:
    clean_country = country.strip()
    if len(clean_country) < 2 or not clean_country.replace(" ", "").isalpha():
        continue
    try:
        location = geocode(clean_country)
        if location:
            country_coords[country] = (location.latitude, location.longitude)
    except:
        continue

m = folium.Map(location=[30, 0], zoom_start=2)
max_count = max(pair_counts.values())

def count_to_color(count, max_count):
    ratio = count / max_count
    red = int(255 * ratio)
    green = int(255 * (1 - ratio))
    return f"#{red:02x}{green:02x}00"

for (c1, c2), count in pair_counts.items():
    if c1 in country_coords and c2 in country_coords:
        coords1, coords2 = country_coords[c1], country_coords[c2]
        color = count_to_color(count, max_count)
        folium.PolyLine(
            [coords1, coords2],
            color=color,
            weight=4,
            opacity=0.8
        ).add_to(m)

m.save("translated_country_mentions_map.html")
print("✔️ Map saved as 'translated_country_mentions_map.html'")

fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>First Country</b>", "<b>Second Country</b>", "<b>Mentions</b>", "<b>Sentiment</b>"],
        fill_color="#264653",
        font=dict(color='white', size=14),
        align='center',
        height=35
    ),
    cells=dict(
        values=[
            pair_df["First Country"],
            pair_df["Second Country"],
            pair_df["Message Count"],
            pair_df["Sentiment Label"]
        ],
        fill_color='lavender',
        align='center',
        font=dict(size=12),
        height=30
    )
)])

fig.update_layout(
    title="Country Pairs with Majority Sentiment from Telegram Messages",
    title_x=0.5,
    margin=dict(l=20, r=20, t=50, b=20),
    width=800,
    height=850
)
fig.show()
```

```python

import pandas as pd
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import umap
import plotly.express as px

df_israel = pd.read_csv("telegram_messages_israel_2019plus_translated.csv")
translated_texts = df_israel['translated_message'].dropna().tolist()

sbert_model = SentenceTransformer("all-MiniLM-L6-v2")
kw_model = KeyBERT(sbert_model)

topics = kw_model.extract_keywords(
    translated_texts,
    keyphrase_ngram_range=(1, 3),
    stop_words='english',
    top_n=1
)

all_keywords = [kw for doc in topics if doc for kw in doc]  # כל kw הוא (מילה, ציון)
keyword_texts = [kw[0] for kw in all_keywords]

keyword_counts = pd.Series(keyword_texts).value_counts().reset_index()
keyword_counts.columns = ["Topic", "Count"]

embeddings = sbert_model.encode(keyword_counts["Topic"].tolist(), convert_to_tensor=False)

n_clusters = 30
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(embeddings)
keyword_counts["Cluster"] = clusters

reducer = umap.UMAP(n_neighbors=10, min_dist=0.3, metric='cosine', random_state=42)
embedding_2d = reducer.fit_transform(embeddings)
keyword_counts["x"] = embedding_2d[:, 0]
keyword_counts["y"] = embedding_2d[:, 1]

representatives = keyword_counts.groupby("Cluster").apply(lambda g: g.nlargest(1, "Count")).reset_index(drop=True)

fig = px.scatter(
    representatives,
    x="x",
    y="y",
    size="Count",
    text="Topic",
    color="Cluster",
    title="Broader Semantic Topics (Clustered & Aggregated)",
    size_max=60
)
fig.update_traces(textposition='top center')
fig.show()
```

This script filters Pikud HaOref Telegram messages for red alert keywords, extracts Hebrew location mentions by month, geocodes them using Nominatim, and saves a monthly alert frequency dataset with coordinates.

```python
import pandas as pd
import re
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter


df = pd.read_csv("pikud_haoref_messages_2019_onward.csv")
df['message'] = df['message'].astype(str)
df['date_parsed'] = pd.to_datetime(df['date'])


alert_keywords = ["עדכון - התרעות", "ירי רקטות", "הופעלה התרעה", "נשמעה אזעקה"]
df_alerts = df[df['message'].apply(lambda x: any(k in x for k in alert_keywords))]

def extract_locations(row):
    message = row['message']
    date = pd.to_datetime(row['date']).to_period('M')
    lines = message.split('\n')
    locations = set()
    for line in lines:
        line = line.strip()
        if any(k in line for k in ["www", "מוקד", "104", "052", ".il", "number", "מספר"]):
            break
        if line.startswith("אזור"):
            continue
        line_cleaned = re.sub(r"\([^)]*\)", "", line)
        parts = [p.strip() for p in re.split(r'[,\n]', line_cleaned)]
        for part in parts:
            if re.fullmatch(r"[א-ת\s]{2,}", part):  # רק טקסט עברי
                locations.add((date, part))
    return list(locations)

mentions = []
for _, row in df_alerts.iterrows():
    mentions.extend(extract_locations(row))

df_mentions = pd.DataFrame(mentions, columns=["month", "location"])


grouped = df_mentions.groupby(["month", "location"]).size().reset_index(name="count")

geolocator = Nominatim(user_agent="red_alert_locator")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

grouped['latitude'] = None
grouped['longitude'] = None

for idx, row in grouped.iterrows():
    try:
        loc = geocode(f"{row['location']}, Israel")
        print(loc)
        if loc:
            grouped.at[idx, 'latitude'] = loc.latitude
            grouped.at[idx, 'longitude'] = loc.longitude
    except Exception:
        continue


grouped.to_csv("full_location_alerts_with_coords.csv", index=False)
```

```python
import pandas as pd
from folium import Map, LayerControl, FeatureGroup, CircleMarker

file_path = "full_location_alerts_flexible.csv"
df = pd.read_csv(file_path)

df_geo = df.dropna(subset=["latitude", "longitude"]).copy()

df_geo["month"] = pd.to_datetime(df_geo["month"])

grouped = df_geo.groupby(["month", "location", "latitude", "longitude"], as_index=False)["count"].sum()
grouped["month_str"] = grouped["month"].dt.strftime("%Y-%m")

m = Map(location=[31.5, 34.75], zoom_start=8, tiles="CartoDB positron")

for month_str, month_df in grouped.groupby("month_str"):
    fg = FeatureGroup(name=month_str, show=False)
    for _, row in month_df.iterrows():
        CircleMarker(
            location=(row["latitude"], row["longitude"]),
            radius=min(20, 3 + row["count"] / 2),
            color="red",
            fill=True,
            fill_opacity=min(0.9, 0.2 + row["count"] / 50),
            popup=f"{row['location']} ({row['count']} alerts)",
        ).add_to(fg)
    fg.add_to(m)

LayerControl(collapsed=False).add_to(m)

output_path = "alerts_map_by_month_dropdown.html"
m.save(output_path)

print(f"✔️ HTML saved to: {output_path}")
```
